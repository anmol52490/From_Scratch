1st we will be implementing a simple 1 neuron with 2 inputs which gives out 1 output.

```python
import numpy as np
```

```python
def sigmoid(x):
    return np.exp(x)/(1+np.exp(x))
```

```python
class Neuron:
    def __init__(self, w, b):
        self.w = w
        self.b = b

    def forward(self,x):
        return sigmoid(np.dot(self.w,x)+self.b)
```

```python
w = np.array([2,3])
b = np.array([1])
n = Neuron(w,b)
x = np.array([2,1])
n.forward(x)
```

So, yea this is it a single neuron operation. Now, Let's take it a step forward with building 2 inputs from input layer 1 hidden layer with 2 hidden neuron and 1 output layer with 1 single neuron with same weight and biases for all of 'em.

```python
class NeuralNetwork:
    def __init__(self):
        self.h1 = Neuron(w,b)
        self.h2 = Neuron(w,b)
        self.o = Neuron(w,b)

    def forward(self, x):
        return self.o.forward(np.array([self.h1.forward(x),self.h2.forward(x)]))
    
nn= NeuralNetwork()
nn.forward(x)
```

Now, that we have done this now let's step a bit forward with Training the neuralnetwork and for that
we will start with Loss

```python
def mse_loss(y_true, y_pred):
    return ((y_true - y_pred)**2).mean()

y_true = np.array([1,0,0,1])
y_label = np.array([0,0,0,0])
mse_loss(y_true, y_label)
```
